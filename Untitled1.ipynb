## Exploratory Data Analysis

```python
df.columns
```

```python
ax = sns.countplot(x ="Size",data=df)
```

```python
print(df.columns)
```

```python
##  remove any leading or trailing spaces from all column 
df.columns =df.columns.str.strip()
```

```python
ax = sns.countplot(x= "Size" ,data=df)
for bars in ax.containers:
    ax.bar_label(bars)
    plt.title("barGraph size of values data")
plt.show()
```

```python
quty = df.groupby(["Size"],as_index=False)["Qty"].sum().sort_values(by="Qty",ascending=False)
sns.barplot(x ="Size" ,y="Qty" ,data=quty)
plt.title("count size")
plt.show()
```

```python
plt.figure(figsize=(10, 12))
ax =sns.countplot(data=df ,x ='Courier Status', hue="Status")
for bars in ax.containers:
    ax.bar_label(bars)
    plt.title("BarGraph of Status")
plt.show()
```

```python
df["Category"] =df["Category"].astype(str)
columns =df["Category"]
plt.figure(figsize=(12,14))
plt.hist(columns, bins=15, edgecolor ="black")
plt.title("BarGraph of categorical_data")
plt.show()
```

```python
Amount_ax = df["B2B"].value_counts()
## plot the pie chart
plt.pie(Amount_ax, labels=Amount_ax, autopct="%1.1f%%")
plt.show()
```

```python
## prepare data for the scatter plot

x_data =df["Category"]
y_data =df["Size"]

plt.scatter(x_data, y_data)
plt.xlabel("Category")
plt.ylabel("Size")
plt.title("Scatter plot")
plt.show()
```

```python
plt.figure(figsize=(12,15))
sns.countplot(data=df, x="ship-state",edgecolor ="black")
plt.xlabel("ship-state")
plt.ylabel("count")
plt.title("Distributions of states")
plt.xticks(rotation =90)
plt.show()
```
